## NETWORK DESCRIPTION

Description:

This project analyzes the social structure of the Marvel Universe character network. It models a "collaboration network" where interactions between fictional characters mimic real-world social dynamics.

Nodes and Edges Representation:
* Nodes: Characters from the Marvel Universe (e.g. Spider-Man, Iron Man).
* Edges: A link connects two nodes if the characters have appeared together in the same comic book issue.
* Network Type: The network is undirected (co-appearance is mutual) and weighted (the weight represents the amount of co-occurrences).



## 1.1 Data downloading and cleaning

Data Source:
[The Marvel Universe Social Network](https://www.kaggle.com/datasets/csanhueza/the-marvel-universe-social-network).

Dataset used:
* hero-network.csv: Contains pairs of heroes that appear in the same comic.

Data Fields Explanation:
* hero1, hero2: String identifiers for the characters.
* weight: The total number of times  hero1 and hero2 appeared together.

To get the data ready for analysis, I need to load the raw edge list, drop any self-loops, and aggregate repeated co-occurrences into a single weighted edge for each character pair.

```python
import pandas as pd
import networkx as nx
import matplotlib.pyplot as plt
import seaborn as sns
import numpy as np

df_edges = pd.read_csv('hero-network.csv')

print("\n--- Raw data sample ---")
display(df_edges.head(5))

#Remove self-loops (Spider-Man connected to Spider-Man) - without this line, there are 12 more total edges in graph - which are artifacts present in raw Kaggle dataset
df_edges = df_edges[df_edges['hero1'] != df_edges['hero2']]

# Remove whitespaces from hero names and sort alphabetically
df_edges['hero1'] = df_edges['hero1'].str.strip()
df_edges['hero2'] = df_edges['hero2'].str.strip()
df_edges[['hero1', 'hero2']] = np.sort(df_edges[['hero1', 'hero2']], axis=1)

# Group by pairs to calculate weights
df_weighted = df_edges.groupby(['hero1', 'hero2']).size().reset_index(name='weight')

print("\n--- Processed weighted data sample ---")
df_weighted.sort_values(by='weight', ascending=False, inplace=True)
display(df_weighted.head(5))

G_raw = nx.from_pandas_edgelist(df_weighted, source='hero1', target='hero2', edge_attr='weight')
print(f"\nGraph created with {G_raw.number_of_nodes()} nodes and {G_raw.number_of_edges()} edges.")
```

## 1.2 Network characterization

I characterize the main parameters of the network, identify and then calculate metrics of the giant component.

```python
nodes_raw = G_raw.number_of_nodes()
edges_raw = G_raw.number_of_edges()
components = list(nx.connected_components(G_raw))

print(f"Network before giant component identification:")
print(f"- Nodes: {nodes_raw}")
print(f"- Edges: {edges_raw}")
print(f"- Number of connected components: {len(components)}")

# Extract giant component
gc_nodes = max(components, key=len)
G = G_raw.subgraph(gc_nodes).copy()

nodes_gc = G.number_of_nodes()
edges_gc = G.number_of_edges()

print(f"\nGiant component:")
print(f"- Nodes: {nodes_gc} ({(nodes_gc/nodes_raw)*100:.2f}% of original)")
print(f"- Edges: {edges_gc} ({(edges_gc/edges_raw)*100:.2f}% of original)")
```

Almost the entire Marvel Universe (over 99% of nodes) is connected within a single giant component. This portrays the publishing strategy of Marvel Comics, which is that the crossovers are frequent and almost every character is connected to the main narrative.

I will use this giant component for all further calculations.

```python
plt.style.use('ggplot')

degrees = [d for n, d in G.degree()]
avg_degree = sum(degrees) / nodes_gc
clustering_coeff = nx.average_clustering(G)
avg_path_length = nx.average_shortest_path_length(G)
diameter = nx.diameter(G)

print(f"Average degree: {avg_degree:.2f}")
print(f"Average clustering coefficient: {clustering_coeff:.4f}")
print(f"Average shortest path length: {avg_path_length:.4f}")
print(f"Network diameter: {diameter}")

# Plot degree distribution
fig, ax = plt.subplots(figsize=(10, 5))

sns.histplot(degrees, bins=50, kde=False, color='steelblue', ax=ax)

ax.set_title('Degree Distribution', fontsize=14)
ax.set_xlabel('Degree', fontsize=12)
ax.set_ylabel('Number of characters', fontsize=12)

plt.tight_layout()
plt.show()
```

The average degree of 52 is quite high, meaning that characters interact with many other, and high clustering coefficient of 0.77 shows that they form teams (like Avengers or X-Men). The average shortest path length of 2.64 shows that the Marvel Universe is a "Small World", which means that any two characters are linked by a very small amount of mutual acquaintances. The degree distribution plot is heavily right-skewed, this means that a majority of characters have very few connections, while the most popular few have hundreds.

## 1.3 Centrality measures

I calculate three centrality measures to identify the most important characters in the network. The graph is undirected, so I use Degree, Closeness, and Betweenness centralities.

```python
# k=200 is used for betweenness to approximate and save computational time
deg_cent = nx.degree_centrality(G)
clo_cent = nx.closeness_centrality(G)
bet_cent = nx.betweenness_centrality(G, k=200, weight='weight', normalized=True, seed=20)

df_centrality = pd.DataFrame({
    'Character': list(G.nodes()),
    'Degree': list(deg_cent.values()),
    'Closeness': list(clo_cent.values()),
    'Betweenness': list(bet_cent.values()),
})

metrics = ['Degree', 'Closeness', 'Betweenness']
for metric in metrics:
    print(f"\n--- Top 5 by {metric} centrality ---")
    display(df_centrality.sort_values(by=metric, ascending=False).head(5)[['Character', metric]])
```

Degree centrality identifies characters with the highest number of collaborators. Captain America, Spider-Man and Iron Man are the faces of the Marvel Comics so they are inserted into many comic series to boost sales. Closeness centrality measures how fast a character can reach anyone else in the network, and the characters with the biggest closeness are the same as those with the largest degree. Characters with high betweenness act as bridges between different teams or storylines (Wolverine, both an X-men and Avengers member) and ensure that the whole universe is connected.

## 1.4 Comparison with network models

To see if the Marvel Universe network is random or if it follows specific patterns, I compare its metrics with three theoretical models: Erdos-Renyi (Random), Watts-Strogatz (Small-World), and Barabasi-Albert (Scale-Free).

```python
n = G.number_of_nodes()
m = G.number_of_edges()

#Random Graph (Erdos-Renyi)
p_er = (2 * m) / (n * (n - 1))
G_er = nx.erdos_renyi_graph(n, p_er, seed=20)

#Small-World Graph (Watts-Strogatz)
k_ws = int(avg_degree)
p_ws = 0.1
G_ws = nx.watts_strogatz_graph(n, k_ws, p_ws, seed=20)

#Scale-Free Graph (Barabasi-Albert)
m_ba = int(avg_degree / 2)
G_ba = nx.barabasi_albert_graph(n, m_ba, seed=20)

models_data = []
models = {'Empirical (Marvel)': G, 'Random': G_er, 'Small-World': G_ws, 'Scale-Free': G_ba}

for name, graph in models.items():
    c = nx.average_clustering(graph)
    models_data.append({'Network': name, 'Clustering Coeff': round(c, 4)})

df_models = pd.DataFrame(models_data)
display(df_models)

#Plot degree distributions
deg_emp = [d for n, d in G.degree()]
deg_er = [d for n, d in G_er.degree()]
deg_ba = [d for n, d in G_ba.degree()]

plt.figure(figsize=(10, 6))
sns.kdeplot(deg_emp, log_scale=True, label='Empirical (Marvel)', linewidth=2)
sns.kdeplot(deg_er, log_scale=True, label='Random (Erdos-Renyi)', linestyle='--')
sns.kdeplot(deg_ba, log_scale=True, label='Scale-Free (Barabasi-Albert)', linestyle=':')

plt.title('Degree Distribution Comparison', fontsize=15)
plt.xlabel('Degree', fontsize=12)
plt.ylabel('Density', fontsize=12)
plt.legend()
plt.show()
```

The table shows that the Marvel clustering coefficient is much higher than the random model. This proves that character interactions are not random, they are organized into specific groups. The density plot shows the random graph has a Poisson-like distribution, which is completely different from the Marvel Universe graph. The Marvel network shares the high clustering of the Small-World. This structure works well with comic universe because it allows authors to kepp localized stories, street level stories while still being able to integrate them into big crossovers. The degree distribution plot demonstrates that the Marvel network aligns closely with the heavy tail of the Scale-Free model. This is an example of preferential attachment: as new characters are made, they are more likely to be introduced alongside already famous heroes rather than unknown characters. This is a deliberate publishing strategy to boost the popularity of new characters.

## 2.1 Community analysis

To understand the internal groups within the Marvel Universe, I will use the Louvain method for community detection. This algorithm is highly efficient for large networks because it optimizes the density of links inside communities compared to links between them. By identifying these communities and looking at their most central characters, we can verify if the algorithm successfully detects communities (like superhero teams).

```python
import networkx.algorithms.community as nx_comm

# Louvain community detection
communities = nx_comm.louvain_communities(G, weight='weight', seed=67)
modularity = nx_comm.modularity(G, communities, weight='weight')

print(f"Number of communities detected: {len(communities)}")
print(f"Modularity score: {modularity:.4f}")

# Sort communities by number of nodes
communities_sorted = sorted(communities, key=len, reverse=True)

print("\n--- Top 5 largest communities and their top 5 central characters ---")
for i, comm in enumerate(communities_sorted[:5]):
    comm_nodes = list(comm)
    # Sort nodes in the community by their degree centrality
    comm_nodes_sorted = sorted(comm_nodes, key=lambda x: deg_cent[x], reverse=True)

    print(f"\nCommunity {i+1} (Size: {len(comm)})")
    print(f"Top characters: {', '.join(comm_nodes_sorted[:5])}")
```

```python
import matplotlib.pyplot as plt
import matplotlib.patches as mpatches
import networkx as nx

# Visualization of the top 5 communities
top_communities = communities_sorted[:5]
nodes_to_plot = []
color_map = []
colors = ['yellow', 'orange', 'red', 'green', 'lightblue']
community_names = ['X-Men', 'Avengers', 'Spider-Man', 'Defenders/Magic', 'Fantastic Four']

# Extract top 20 nodes from each of the community
for idx, comm in enumerate(top_communities):
    comm_nodes_sorted = sorted(list(comm), key=lambda x: deg_cent[x], reverse=True)[:20]
    nodes_to_plot.extend(comm_nodes_sorted)
    color_map.extend([colors[idx]] * 20)

# Draw graph
G_sub = G.subgraph(nodes_to_plot)

plt.figure(figsize=(14, 10))
pos = nx.spring_layout(G_sub, k=0.3, seed=67)

nx.draw_networkx_edges(G_sub, pos, alpha=0.2, edge_color='gray')
nx.draw_networkx_nodes(G_sub, pos, node_size=300, node_color=color_map, edgecolors='black', linewidths=0.5)

# Legend
legend_patches = []
for color, name in zip(colors, community_names):
    patch = mpatches.Patch(color=color, label=name)
    legend_patches.append(patch)

leg = plt.legend(handles=legend_patches, title="Top Marvel Communities", fontsize=11, title_fontsize=13,labelcolor = 'black', loc='upper right')

leg.get_title().set_color('black')

plt.title("Core Structure of the Top 5 Marvel Communities", fontsize=20,color='black')
plt.axis('off')
plt.tight_layout()
plt.show()
```

The Louvain method detected 28 communities with a high modularity score of 0.4999. This score indicates a very strong and distinct community structure. By inspecting the top characters in the 5 largest communities (sorted by degree centrality), we can clearly identify the meaning behind them. They align with the most popular groups in Marvel comics:
* Community 1 (The X-Men): Centered around mutants like Wolverine, Beast, Cyclops, and Storm.
* Community 2 (The Avengers): Centred around Captain America, Iron Man, Scarlet Witch, and Vision.
* Community 3 (The Spider-Man & Street-level): Centered around Spider-Man, Daredevil, and Peter Parker's supporting cast (Mary Jane, J. Jonah Jameson).
* Community 4 (The Defenders / Magic): Centered around Dr. Strange, Hulk, and their collaborators like Wong and Rick Jones.
* Community 5 (The Fantastic Four): Centered around Thing, Mr. Fantastic, Human Torch, Invisible Woman, and the Silver Surfer.

This confirms that the network's community structure is highly correlated with Marvel publishing strategy. Characters mainly interact within their established thematic groups (comic book series), with the most central nodes acting as anchors that keep the communities connected to the Marvel Universe.

## 2.2 Network dynamics

For the network dynamics analysis, I will evaluate the robustness of the Marvel Universe.

I will simulate two scenarios and track the size of the giant component:
* Random Failures: Characters are removed at random.
* Targeted Attacks: Characters are removed strategically, starting from the highest degree centrality.

```python
import random

def simulate_removal(graph, nodes_to_remove, steps=40):
    G_sim = graph.copy()
    num_nodes_total = G_sim.number_of_nodes()
    batch_size = num_nodes_total // steps

    fractions_removed = [0]
    gc_sizes = [1.0]

    for i in range(steps):
        # Select batch of nodes to remove
        batch = nodes_to_remove[i*batch_size : (i+1)*batch_size]
        G_sim.remove_nodes_from(batch)

        # Calculate new giant component size
        if G_sim.number_of_nodes() > 0:
            components = list(nx.connected_components(G_sim))
            if components:
                largest_cc = len(max(components, key=len))
                gc_sizes.append(largest_cc / num_nodes_total)
            else:
                gc_sizes.append(0)
        else:
            gc_sizes.append(0)

        fractions_removed.append((i+1)/steps)

    return fractions_removed, gc_sizes

# 1. Random Failures
nodes_random = list(G.nodes())
random.seed(67)
random.shuffle(nodes_random)
frac_rand, gc_rand = simulate_removal(G, nodes_random)

# 2. Targeted Attacks (Based on degree centrality)
nodes_targeted = sorted(deg_cent, key=deg_cent.get, reverse=True)
frac_targ, gc_targ = simulate_removal(G, nodes_targeted)

# Plot
plt.figure(figsize=(10, 6))
plt.plot(frac_rand, gc_rand, label='Random Failures', color='green', linewidth=2.5, marker='o', markevery=4)
plt.plot(frac_targ, gc_targ, label='Targeted Attacks', color='red', linewidth=2.5, marker='o', markevery=4)

plt.title('Network Robustness: Giant Component Size under Node Removal', fontsize=15, color="black")
plt.xlabel('Fraction of Nodes Removed', fontsize=12, color='black')
plt.ylabel('Fraction of Nodes in Giant Component', fontsize=12, color='black')
plt.grid(True, linestyle='--', alpha=0.7)
plt.legend(fontsize=12, labelcolor='black', loc='upper right')
plt.tight_layout()
plt.show()
```

The plot reveals a classic property of scale-free networks.
* Under Random Failures (Green line): The network is highly robust. Even if we remove 40% of random characters, the giant component remains largely intact. This is because most characters in the Marvel Universe have very few connections (as seen in the degree distribution in Part 1), so their removal barely affects the global structure.
* Under Targeted Attacks (Red line): The network is fragile. Removing most of the central characters causes the giant component to rapidly collapse into disconnected components.

In a real-world comic scenario, if Marvel were to stop publishing stories with their most popular heroes, the "Universe" structure would quickly disintegrate, leaving behind only disjointed storylines.

## 2.3 Conclusions and assessment

The analysis of the Marvel Universe Social Network reveals a highly organized and distinct structure that perfectly mirrors its Marvel Comics' publishing strategy:

* Topological Properties: The network exhibits classic "Small-World" properties (short average path lengths, high clustering) and a "Scale-Free" degree distribution. This means the universe relies on a few massive hubs (most popular heroes) to connect a vast number of minor characters.
* Centrality and Influence: Characters with the highest centrality (Captain America, Spider-Man, Iron Man) tie the whole universe together. Their high betweenness centrality acts as bridges, allowing crossovers between different storylines.

* Vulnerability: The dynamic simulation proved that the Marvel Universe is robust to random character disappearances but highly vulnerable to the loss of its core hubs.

Marvel Universe Social Network behaves exactly like a complex real-world social network. The Marvel publishers engineered a highly clustered, scale-free structure that maximizes reader engagement by keeping communities localized (for dedicated readers of specific comics) while maintaining a giant component through massive crossover events.
